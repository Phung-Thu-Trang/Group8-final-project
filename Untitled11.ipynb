```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc file CSV
df = pd.read_csv("IBM_clean.csv")

# Chuẩn hóa tên cột và xử lý thời gian
df.columns = df.columns.str.strip().str.lower()
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Tính daily return và volatility (rolling window 20 ngày)
df['daily_return'] = df['close'].pct_change()
df['volatility'] = df['daily_return'].rolling(window=20).std()

# Loại bỏ giá trị NaN
df_plot = df[['daily_return', 'volatility']].dropna()

# Tính trung vị để phân vùng
return_median = df_plot['daily_return'].median()
vol_median = df_plot['volatility'].median()

# Tạo cột phân vùng
conditions = [
    (df_plot['daily_return'] >= return_median) & (df_plot['volatility'] < vol_median),
    (df_plot['daily_return'] < return_median) & (df_plot['volatility'] >= vol_median),
]
choices = ['High Return – Low Risk', 'Low Return – High Risk']
df_plot['zone'] = np.select(conditions, choices, default='Other')

# Vẽ biểu đồ
plt.figure(figsize=(12, 7))
palette = {'High Return – Low Risk': 'green',
           'Low Return – High Risk': 'red',
           'Other': 'orange'}

sns.scatterplot(data=df_plot, x='volatility', y='daily_return', hue='zone', palette=palette, alpha=0.7)
plt.axhline(return_median, color='blue', linestyle='--', label='Median Return')
plt.axvline(vol_median, color='gray', linestyle='--', label='Median Volatility')

plt.title("Biểu đồ Return vs Volatility (IBM) – Phân vùng rủi ro/lợi nhuận")
plt.xlabel("Volatility (Rolling Std – 20 ngày)")
plt.ylabel("Daily Return")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
